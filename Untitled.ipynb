```python
import pandas as pd
import numpy as np
import scipy as sps


from sklearn.metrics import roc_auc_score

def auc_weighted(y_true, y_pred):
    score = roc_auc_score(y_true, y_pred, average='weighted')
    return score


traindata = pd.read_csv("input_train.csv", delimiter=",")
output = pd.read_csv("output_train.csv", delimiter=";")
input_test = pd.read_csv("input_test.csv", delimiter=",")
```

```python
traindata['SHIPPING_PRICE'] = traindata['SHIPPING_PRICE'].map({'<1': 0, '1<5':1, '5<10':2, '10<20':3, '>20': 4})
traindata['SHIPPING_PRICE'] = traindata['SHIPPING_PRICE'].fillna(0)

traindata['WARRANTIES_PRICE'] = traindata['WARRANTIES_PRICE'].map({'<5': 1, '5<20':2, '20<50':3, '50<100':4, '100<500':5})
traindata['WARRANTIES_PRICE'] = traindata['WARRANTIES_PRICE'].fillna(0)

traindata['PRICECLUB_STATUS'] = traindata['PRICECLUB_STATUS'].fillna("UNSUBSCRIBED")

traindata['PURCHASE_COUNT'] = traindata['PURCHASE_COUNT'].map({'<5':0, '5<20': 1, '20<50':2, '50<100':3, '100<500': 4, '>500':5})

traindata['BUYING_DATE'] = traindata['BUYING_DATE'].map({'1/2017':1, '2/2017':2,'3/2017':3,'4/2017':4,'5/2017':5, '6/2017':6, '7/2017':7, '8/2017':8, '9/2017':9, '10/2017':10})

traindata['SELLER_SCORE_COUNT'] = traindata['SELLER_SCORE_COUNT'].map({'<100':0, '100<1000':1, '1000<10000':2, '10000<100000':3, '100000<1000000':4})
traindata['SELLER_SCORE_COUNT'] = traindata['SELLER_SCORE_COUNT'].fillna(0)

traindata['SELLER_SCORE_AVERAGE'] = traindata['SELLER_SCORE_AVERAGE'].fillna(44)

traindata['ITEM_PRICE'] = traindata['ITEM_PRICE'].map({'<10':0, '10<20':1, '20<50':2, '50<100':3, '100<500':4, '500<1000':5, '1000<5000':6,'>5000':7})
```
